# Task 3: Product Pricing

```python
price_dict = {
    "Laptop": 75000,
    "Phone": 40000,
    "Headphones": 3000,
    "Keyboard": 1500,
    "Mouse": 800,
    "Monitor": 12000
}
```

```python
price_dict["Tablet"] = 25000
price_dict['Laptop'] = 72000
```

```python
product_to_remove = "Camera"
if product_to_remove in price_dict:
    del price_dict[product_to_remove]
```

```python
avg_price = sum(price_dict.values()) / len(price_dict)
print("Average price:", avg_price)
```

```python
max_product = max(price_dict, key=price_dict.get)
min_product = min(price_dict, key=price_dict.get)
print("Max price product:", max_product)
print("Min price product:", min_product)
```

# Task 4: Combined Operations

```python
catalog = [
    ("Laptop", price_dict["Laptop"], "Electronics"),
    ("Phone", price_dict["Phone"], "Electronics"),
    ("Headphones", price_dict["Headphones"], "Accessories"),
    ("Keyboard", price_dict["Keyboard"], "Accessories"),
    ("Mouse", price_dict["Mouse"], "Accessories"),
    ("Monitor", price_dict["Monitor"], "Electronics"),
]
```

```python
category_to_products = {}
for name, _, category in catalog:
    category_to_products.setdefault(category, []).append(name)
```

```python
max_category = max(category_to_products, key=lambda c: len(category_to_products[c]))
print("Category with max products:", max_category)
print("Products:", category_to_products[max_category])
```
